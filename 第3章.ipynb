```
import json
import codecs
```

## 20

```
data = []
with codecs.open("jawiki-country.json", 'r', encoding='utf_8', errors="ignore") as f:
  for i in f:
    try:
      data.append(json.loads(i))
    except:
      pass
```

```
uk = list(filter(lambda x: x['title'] == "イギリス", data))[0]
```

## 21

```
import re
```

```
cate = re.findall(r"\[\[Category\:.+?\]\]", uk['text'])
```

```
cate
```

## 22

```
cate = re.finditer(r"\[\[Category\:(?P<category_name>.+?)(\|.+)?\]\]", uk['text'])
```

```
for i in cate:
  print(i.group('category_name'))
```

## 23

```
res = []
for text in uk['text'].split("\n"):
  m = re.match(r"(?P<level>=+)(?P<name>.+?)=+", text)
  if m:
    res.append(m)
```

```
for i in res:
  print(i.group(), len(i.group("level")) - 1)
```

## 25

```
depth = [0 for _ in range(len(uk['text']))]
```

```
for i,t in enumerate(uk['text']):
  depth[i] = depth[i-1]
  if uk['text'][i:i+2] == '{{':
    depth[i] = depth[i-1] + 1
  if i > 1 and uk['text'][i-2:i] == '}}':
    depth[i] = depth[i-1] - 1
```

```
start = re.search(r"{{基礎情報", uk['text'])
print(start.start())
print(depth[start.start()])
```

```
s_depth = depth[start.start()]
for i in range(start.start(), len(uk['text'])):
  if depth[i] < s_depth:
    break

info = uk['text'][start.start():i]
```

```
m = re.finditer(r"\|(?P<key>.+?)=(?P<value>.+)", info)
```

```
ans = {i.group('key').strip(): i.group('value').strip() for i in m}
# 複数行のやつとれないから厳密には違うかも？いろんなパターンに対応するのめんどくさいからとりあえずこれで
```

## 26

```
for key in ans:
  ans[key] = re.sub(r"''+(.+?)''+", r"\1", ans[key])
```

```
ans
```
