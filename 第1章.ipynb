## 05

```
S = "I am an NLPer"

def n_gram(elements, n):
  ret = [elements[i:i+n] for i in range(len(elements)-n+1)]
  return ret

print(n_gram(S, 2))
print(n_gram(S.split(), 2))
```

## 06

```
S1 = "paraparaparadise"
S2 = "paragraph"

bi_1 = set(n_gram(S1, 2))
bi_2 = set(n_gram(S2, 2))

print(bi_1)
print(bi_2)

print(bi_1 | bi_2)
print(bi_1 & bi_2)
print(bi_1 - bi_2)

print("se" in bi_1)
print("se" in bi_2)
```

## 08

```
def cipher(sentence):
  res = []
  for c in sentence:
    if 97 <= ord(c) < 97+26:
      res.append(chr(219 - ord(c)))
    else:
      res.append(c)
  return "".join(res)
  

S = "Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."
print(S)
c_s = cipher(S)
print(c_s)
de_c_s = cipher(c_s)
print(de_c_s)
```

## 09

```
import random
S = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind."

def shuffle(char):
  if len(char) < 4:
    return char
  shuffled = char[0] + "".join(random.sample(list(char[1:-1]), len(char)-2)) + char[-1]
  return shuffled

def proc(S):
  return " ".join([shuffle(c) for c in S.split()])

print(proc(S))
```
